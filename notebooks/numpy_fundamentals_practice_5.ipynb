```python
# Array Concatenation and Splitting
import numpy as np
a= np.array([1,2,4,6])
b= np.array([1,26,46,67])
c=np.concatenate((a,b))
print("concatenated =",c)
```

```python
g = np.array([[1, 2], [3, 4]])
m = np.array([[5, 6]])
cp=np.concatenate((g,m),axis=0)
print("concatenated2 =",cp)
```

```python
o = np.array([[1, 45],[3, 4]])
l = np.array([[10],[20]])
ol=np.concatenate((o,l),axis=1)
print("concatenated3 =",ol)
```

```python
k = np.array([1, 2])
s = np.array([3, 4])
sk=np.vstack((k,s))
ks=np.hstack((k,s))
print("vstack =",sk)
print("hstack =",ks)
arr = np.array([10, 20, 30, 40, 50, 60])
splitarr=np.split(arr,2)
print("split =",splitarr)
```

```python
# splitting 2d array
matrix = np.array([[1, 2, 3, 4],[5, 6, 7, 8]])
top,bottom=np.vsplit(matrix,2)
print("top =",top)
print("bottom =",bottom)
left,right=np.hsplit(matrix,2)
print("left =",left)
print("right=",right)
```

```python
# elementwise arithmetic operations
import  numpy as np
a = np.array([1,2,3,4,5])
b = np.array([1,5,89,4,7])
print("sum =",a+b)
print("product =",a*b)
print("division =",a/b)
print("exponent =",a**b)
print("modulo =",a%b)
# scalar operations
print(a+3)
print(a**3)
print(a/3)
```

```python
# comparison operations
a = np.array([1,2,3,4,5])
print(a>=3)
print(a<3)
print(a==3)
# logical operations
arr = np.array([33,44,55,67,78])
print((arr>70)| (arr<60))
print((arr>70) & (arr<767))
print((arr==70) | (arr<767))
print((arr==70) | (arr>767))
```

```python
# aggregate functions
arr = np.array([33,44,55,67,78])
print(arr.min())
print(arr.max())
print(arr.mean())
print(arr.var())
print(arr.sum())
print(arr.std())
```

```python
# axis based aggregations
arr = np.array([[80, 90, 70],[60, 85, 75]])
print(arr.sum(axis=0))
print(arr.sum(axis=1))
print(arr.min(axis=1))
print(arr.mean(axis=1))
```

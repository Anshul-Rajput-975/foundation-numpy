```python
# use of copy in numpy
import numpy as np
a= np.array([1,2,3,4,5,6])
print("a =",a)
b=a.copy()
b[1]=4567
print("a =",a) # no change in a as copy results in creation of new memory
print("b =",b)
```

```python
# use of view in short in numpy
import numpy as np
a= np.array([1,2,3,4,5,6])
print("a =",a)
b=a.reshape(2,3)
b[(0,2)]=44
print("a = ",a) # change in a as view results in creation of shared memory

print("b = ",b)
```

```python
 # indexing and slicing
import numpy as np
arr=np.arange(9)
print(arr)
print(arr[:8])
print(arr[3:6])
print(arr[8])
print(arr[1:8:2])
print(arr[1:8:3])
print(arr[1::3])
print(arr[8:0:-1])
```

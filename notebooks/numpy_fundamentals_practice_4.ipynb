```python
# broadcasting in numpy
import numpy as np
c=np.array([[50, 60, 70],[80, 90, 100]])
b=12
curve = np.array([1, 2, 3])
print("sum =",c+b)
print("power =",c**b)
print("product =",c*curve)
print("subtraction =",c-curve)
```

```python
# reshaping and resizing arrays
curve = np.array([1, 2, 3,4,5,6,7,8])
b=curve.reshape(4,2)
print("b =",b)
c=b.flatten()
print("c =", c)
b=np.resize(b,(5,5))
print("b =",b)
b=b.reshape(5,5)
print("b =",b)
```

```python
# NUMPY AND LINEAR ALGEBRA
import numpy as np
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
print(np.dot(A,B))
print(A*B)
print(np.transpose(A))
c= np.array([[1, 2], [3, 4]])
```

```python
import numpy as np
c= np.array([[1, 2], [3, 4]])
det=np.linalg.det(c)
inver=np.linalg.inv(c)
print(det)
print(inver)
A = np.array([[4, -2], [1, 1]])
eigvals, eigvecs = np.linalg.eig(A)
print("Eigenvalues =", eigvals)
print("Eigenvectors =", eigvecs)
```

```python
# for solving equation AX=b
A = np.array([[2, 1], [1, 3]])
b = np.array([8, 18])
X=np.linalg.solve(A,b)
print("value of x =", X)
```

```python
#  Numpy Handling Missing Data
data = np.array([1, 2, np.nan, 4, 5, np.nan]) # basically np.nan is written for handling of missing values
print("data =",data)
```

```python
#checking for Missing Data (np.isnan())
data = np.array([1, 2, np.nan, 4, 5, np.nan])
is_nan=np.isnan(data)
print("is_nan =",is_nan)
```

```python
#Replacing Missing Values (np.nan_to_num())
data = np.array([1, 2, np.nan, 4, 5, np.nan])
is_data=np.nan_to_num(data)
print("is_data =",is_data)
```

```python
#Filling Missing Values with a Specific Value (np.where())
data = np.array([1, 2, np.nan, 4, 5, np.nan])
where_data=np.where(data)
print("where_data =",where_data)
```

```python
#Removing Missing Data (np.isnan())
data = np.array([1, 2, np.nan, 4, 5, np.nan])
cleaneddata=data[~np.isnan(data)]
print("cleaneddata =",cleaneddata)
```

```python
#Using np.nanmean(), np.nanstd(), etc.
# Using np.nanmean() to compute the mean while ignoring NaN
data = np.array([1, 2, np.nan, 4, 5, np.nan])
mean=np.nanmean(data)
std=np.nanstd(data)
print("mean =",mean)
print("std =",std)
```

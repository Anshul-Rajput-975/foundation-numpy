```python
 # Iterating Over Arrays
import numpy as np
fruits = np.array(['apple', 'banana', 'cherry'])
for fruit in fruits:
    print("fruits=",fruit)
```

```python
arr = np.array([[1, 2], [3, 4], [5, 6]])
for row in arr:
    print("rows =",row)
```

```python
arr = np.array([[1, 2], [3, 4]])
for val in arr.flat:
    print(val, end=" ")
```

```python
arr = np.array([1, 2, 3,4])
#p.nditer is used to loop through each element of the array one by one.
#op_flags=['readwrite'] allows modifying the array values while iterating.
for x in np.nditer(arr, op_flags=['readwrite']):
    x[...]=x**3  
print("new =",arr)    
```

```python
# np.ndenumerate() gives both index and value, which is helpful in debugging
arr = np.array([[100, 200], [300, 400]])
for index,value in np.ndenumerate(arr):
    print(f"index is {index}",f" value is {value}")
```

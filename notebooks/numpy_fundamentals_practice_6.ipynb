##copying and views in numpy

```python
# crating view using slicing
import numpy as np
arr = np.array([10, 20, 30, 40])
print("arr =",arr)
view=arr[1:]
print("view =",view)
view[0]=567
print("view =",view)
print("arr =",arr)
# Creating a Copy 
ar = np.array([10, 200, 308, 40])
print("ar =",ar)
b=np.copy(ar)
print("b =",b)
b[0]=567
print("b =",b)
print("ar =",ar)
# Checking if Arrays Share Memory with h np.may_share_memory()
from numpy import may_share_memory
print(may_share_memory(arr,view))
print(may_share_memory(ar,b))
```
